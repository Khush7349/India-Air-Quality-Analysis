```python
import pandas as pd
import matplotlib.pyplot as plt
files_to_analyze = [
    '/kaggle/input/time-series-air-quality-data-of-india-2010-2023/AP001.csv', 
    '/kaggle/input/time-series-air-quality-data-of-india-2010-2023/DL001.csv', 
    '/kaggle/input/time-series-air-quality-data-of-india-2010-2023/MH005.csv', 
    '/kaggle/input/time-series-air-quality-data-of-india-2010-2023/WB013.csv'  
]
all_station_data = []
for file in files_to_analyze:
    station_name = file.split('/')[-1].replace('.csv', '')
    print(f"Processing station: {station_name}")
    df_temp = pd.read_csv(file)
    df_temp['From Date'] = pd.to_datetime(df_temp['From Date'])
    df_temp.set_index('From Date', inplace=True)
    df_temp.interpolate(method='linear', inplace=True)
    df_temp['Station'] = station_name
    all_station_data.append(df_temp)
combined_df = pd.concat(all_station_data)
print("\n--- Combined Data Info ---")
combined_df.info()
print("\n--- Combined Data Head ---")
display(combined_df.head())
```

```python
plt.figure(figsize=(17, 8))
stations = combined_df['Station'].unique()
for station in stations:
    station_df = combined_df[combined_df['Station'] == station]
    monthly_avg = station_df['PM2.5 (ug/m3)'].resample('M').mean()
    plt.plot(monthly_avg.index, monthly_avg, label=station, marker='o', markersize=2, linestyle='--')
plt.title('Monthly Average PM2.5 Comparison Across Cities', fontsize=16)
plt.xlabel('Date')
plt.ylabel('Average PM2.5 Concentration (ug/m3)')
plt.grid(True)
plt.legend(title='Station ID') 
plt.show()
```

```python
import pandas as pd
import glob
all_files = sorted([f for f in glob.glob('/kaggle/input/time-series-air-quality-data-of-india-2010-2023/*.csv') if 'info' not in f])
files_to_analyze = all_files[:30] 
print(f"Found {len(files_to_analyze)} files to process.")
all_station_data_full = []
for file in files_to_analyze:
    station_name = file.split('/')[-1].replace('.csv', '')
    try:
        df_temp = pd.read_csv(file)
        if 'From Date' in df_temp.columns:
            df_temp['From Date'] = pd.to_datetime(df_temp['From Date'], errors='coerce')
            df_temp.dropna(subset=['From Date'], inplace=True)
            df_temp.set_index('From Date', inplace=True)
            df_temp.interpolate(method='linear', limit_direction='both', inplace=True) 
            df_temp['Station'] = station_name
            all_station_data_full.append(df_temp) 
        else:
            print(f"Skipping {station_name}: missing 'From Date' column.")
    except Exception as e:
        print(f"Could not process {station_name}. Error: {e}")
combined_df_full = pd.concat(all_station_data_full)
print("\n--- New Combined DataFrame Info ---")
combined_df_full.info()
```

```python
import seaborn as sns
import matplotlib.pyplot as plt
def plot_correlation_heatmap(station_id, df):
    print(f"--- Correlation Heatmap for Station: {station_id} ---")
    station_df = df[df['Station'] == station_id].copy()
    columns_for_corr = ['PM2.5 (ug/m3)', 'PM10 (ug/m3)', 'NO2 (ug/m3)', 'CO (mg/m3)', 'Ozone (ug/m3)', 'SO2 (ug/m3)', 'Temp (degree C)', 'RH (%)', 'WS (m/s)']
    columns_for_corr = [col for col in columns_for_corr if col in station_df.columns] 
    if len(columns_for_corr) < 2:
        print(f"Not enough data to create a heatmap for {station_id}")
        return
    correlation_matrix = station_df[columns_for_corr].corr()
    plt.figure(figsize=(10, 7))
    sns.heatmap(correlation_matrix, annot=True, cmap='coolwarm', fmt='.1f', linewidths=.5)
    plt.title(f'Correlation Matrix for {station_id}', fontsize=14)
    plt.show()
plot_correlation_heatmap('DL039', combined_df_full)
plot_correlation_heatmap('AP001', combined_df_full)
```

```python
import pandas as pd
import glob
import matplotlib.pyplot as plt
import seaborn as sns
all_files = sorted([f for f in glob.glob('/kaggle/input/time-series-air-quality-data-of-india-2010-2023/*.csv') if 'info' not in f])
files_to_analyze = all_files[:30]
all_station_data = []
for file in files_to_analyze:
    station_name = file.split('/')[-1].replace('.csv', '')
    try:
        df_temp = pd.read_csv(file)
        if 'From Date' in df_temp.columns:
            df_temp['From Date'] = pd.to_datetime(df_temp['From Date'], errors='coerce')
            df_temp.dropna(subset=['From Date'], inplace=True)
            df_temp.set_index('From Date', inplace=True)
            df_temp.interpolate(method='linear', limit_direction='both', inplace=True)
            df_temp['Station'] = station_name
            all_station_data.append(df_temp)
    except Exception as e:
        print(f"Skipping {station_name}, error: {e}")
combined_df_full = pd.concat(all_station_data)
station_summary = combined_df_full.groupby('Station')['PM2.5 (ug/m3)'].mean().sort_values(ascending=False).reset_index()
station_summary.rename(columns={'PM2.5 (ug/m3)': 'Average PM2.5'}, inplace=True)
print("\n--- Generating Dashboard ---")
fig, ax = plt.subplots(2, 2, figsize=(20, 15))
fig.suptitle('Air Quality Analysis Dashboard: Key Insights', fontsize=20, weight='bold')
top_10_polluted = station_summary.head(10)
sns.barplot(x='Average PM2.5', y='Station', data=top_10_polluted, ax=ax[0, 0], palette='Reds_r')
ax[0, 0].set_title('Top 10 Most Polluted Stations', fontsize=14, weight='bold')
stations_to_plot = ['BR008', 'AP001', 'MH005']
for station in stations_to_plot:
    if station in combined_df_full['Station'].unique():
        monthly_avg = combined_df_full[combined_df_full['Station'] == station]['PM2.5 (ug/m3)'].resample('ME').mean()
        ax[0, 1].plot(monthly_avg.index, monthly_avg, label=station)
ax[0, 1].set_title('Monthly PM2.5 Levels: High vs. Low Pollution', fontsize=14, weight='bold')
ax[0, 1].legend()
station_for_heatmap = 'BR008'
if station_for_heatmap in combined_df_full['Station'].unique():
    columns_for_corr = ['PM2.5 (ug/m3)', 'PM10 (ug/m3)', 'NO2 (ug/m3)', 'CO (mg/m3)', 'Temp (degree C)', 'WS (m/s)']
    columns_for_corr = [col for col in columns_for_corr if col in combined_df_full.columns]
    correlation_matrix = combined_df_full[combined_df_full['Station'] == station_for_heatmap][columns_for_corr].corr()
    sns.heatmap(correlation_matrix, annot=True, cmap='coolwarm', fmt='.1f', ax=ax[1, 0])
ax[1, 0].set_title(f'Pollutant & Weather Correlation ({station_for_heatmap})', fontsize=14, weight='bold')
top_6_stations = station_summary['Station'].head(6).tolist()
boxplot_df = combined_df_full[combined_df_full['Station'].isin(top_6_stations)]
sns.boxplot(x='Station', y='PM2.5 (ug/m3)', data=boxplot_df, ax=ax[1, 1], palette='plasma')
ax[1, 1].set_title('Distribution of PM2.5 in Top Polluted Stations', fontsize=14, weight='bold')
ax[1, 1].set_ylim(0, 500)
plt.tight_layout(rect=[0, 0.03, 1, 0.95])
plt.savefig('air_pollution_dashboard.png', dpi=300)
plt.show()
```

```python
import pandas as pd
import glob
import matplotlib.pyplot as plt
import seaborn as sns
all_files = sorted([f for f in glob.glob('/kaggle/input/time-series-air-quality-data-of-india-2010-2023/*.csv') if 'info' not in f])
files_to_analyze = all_files[:30]
all_station_data = []
for file in files_to_analyze:
    station_name = file.split('/')[-1].replace('.csv', '')
    try:
        df_temp = pd.read_csv(file)
        if 'From Date' in df_temp.columns:
            df_temp['From Date'] = pd.to_datetime(df_temp['From Date'], errors='coerce')
            df_temp.dropna(subset=['From Date'], inplace=True)
            df_temp.set_index('From Date', inplace=True)
            df_temp.interpolate(method='linear', limit_direction='both', inplace=True)
            df_temp['Station'] = station_name
            all_station_data.append(df_temp)
    except Exception as e:
        print(f"Skipping {station_name}, error: {e}")
combined_df_full = pd.concat(all_station_data)
station_summary = combined_df_full.groupby('Station')['PM2.5 (ug/m3)'].mean().sort_values(ascending=False).reset_index()
print("\n--- Generating Final Multi-Pollutant Dashboard ---")
fig, ax = plt.subplots(2, 2, figsize=(22, 12))
fig.suptitle('Air Pollutant Comparison Across Selected Indian Stations', fontsize=20, weight='bold')
pollutants_to_plot = {
    'PM10 (ug/m3)': ax[0, 0],
    'NO2 (ug/m3)': ax[0, 1],
    'CO (mg/m3)': ax[1, 0],
    'Ozone (ug/m3)': ax[1, 1]
}
top_2 = station_summary['Station'].head(2).tolist()
bottom_2 = station_summary['Station'].tail(2).tolist()
middle_index = len(station_summary) // 2
middle_station = [station_summary['Station'].iloc[middle_index]]
stations_to_plot = top_2 + middle_station + bottom_2
print(f"Comparing stations: {stations_to_plot}")
for pollutant, axis in pollutants_to_plot.items():
    if pollutant not in combined_df_full.columns:
        axis.text(0.5, 0.5, f'{pollutant} data not available', horizontalalignment='center', verticalalignment='center')
        axis.set_title(pollutant, fontsize=14, weight='bold')
        continue
    for station in stations_to_plot:
        if station in combined_df_full['Station'].unique():
            monthly_avg = combined_df_full[combined_df_full['Station'] == station][pollutant].resample('ME').mean()
            axis.plot(monthly_avg.index, monthly_avg, label=station)
    axis.set_title(f'Monthly Average {pollutant}', fontsize=14, weight='bold')
    axis.set_ylabel('Concentration')
    axis.set_xlabel('Date')
    axis.legend()
    axis.grid(True)
plt.tight_layout(rect=[0, 0.03, 1, 0.95])
plt.savefig('final_multi_pollutant_dashboard.png', dpi=300)
plt.show()
```
